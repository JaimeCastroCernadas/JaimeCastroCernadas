Las columnas de síntomas y tratamiento, como puede haber muchas combinaciones, vamos a crear una columna para cada síntoma y tratamiento, y le daremos valores 1 y 0

Síntomas


```python
import pandas as pd

def create_and_update_symptom_columns(datos, column_name, symptom_categories):
    for symptom in symptom_categories:
        datos[symptom] = 0
    
    # Function to update the new columns based on the classified symptom values
    def update_columns(row):
        if not pd.isna(row[column_name]):
            for symptom in symptom_categories:
                if symptom in row[column_name]:
                    row[symptom] = 1
        return row
    
    # Apply the function to update the rows
    datos = datos.apply(update_columns, axis=1)
    return datos

# Define symptom categories for each classified symptom column
symptom_mappings = {
    'Tegumentario_clasificado': ['Prurito y Eritema', 'Angioedema/Edema', 'Urticaria', "Otros_tegumentario"],
    'Digestivo_clasificado': ['Dolor abdominal-Náuseas', 'Vómitos-Diarrea', 'Disfagia', "Otros_digestivo"],
    "Respiratorio_clasificado": ["Tos-Estornudos", "Disnea-Broncoespasmo", "Sibilancias", "Dolor torácico", "Otros_respiratorio"],
    'Vascular_clasificado': ['Hipotensión', 'Taquicardia', 'Síncope', "Otros_vascular"],
    'Nervioso_clasificado': ['Mareos', 'Dolor de cabeza', 'Disestesia-Parestesia', 'Dolor lumbar', "Otros_nervioso"]
}

# Assuming `datos` is your DataFrame
# Apply the function for each symptom column
for column, categories in symptom_mappings.items():
    datos = create_and_update_symptom_columns(datos, column, categories)

# Display the first few rows to confirm the changes
print(datos.head())
```

```python
def create_and_update_treatment_columns(datos, column_name, treatment_categories):
    for treatment in treatment_categories:
        datos[treatment] = 0
    
    # Function to update the new columns based on the classified symptom values
    def update_columns(row):
        if not pd.isna(row[column_name]):
            for treatment in treatment_categories:
                if treatment in row[column_name]:
                    row[treatment] = 1
        return row
    
    # Apply the function to update the rows
    datos = datos.apply(update_columns, axis=1)
    return datos
 
    #Define symptom categories for each classified symptom column
treatment_mappings = {
    'Tratamiento_clasificado': ['Corticoesteroides', 'Epinefrina', 'Antagonistas HIST 1-2',"Agonistas B2 adrenérgicos"],
    
}


# Apply the function for each symptom column
for column, categories in treatment_mappings.items():
    datos = create_and_update_treatment_columns(datos, column, categories)

# Display the first few rows to confirm the changes
datos.head()
```

```python
datos.drop(columns=["Tratamiento_clasificado"], inplace=True)
```

Para reacción_clasificado, como sus valores serán mas de 1 y 0, usaremos el One Hot Encoder para evitar que reciban un peso mayor las categorías de mayor valor numérico. En realidad es similar a lo que he hecho con los síntomas, pero con otro método.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Rellena los valores NaN en 'Reaccion_clasificado' con una cadena única
datos['Reaccion_clasificado'] = datos['Reaccion_clasificado'].fillna('missing')

# Inicializa el OneHotEncoder
encoder = OneHotEncoder(sparse=False, drop='if_binary')

# Aplica el OneHotEncoder a la columna 'Reaccion_clasificado'
encoded_data = encoder.fit_transform(datos[['Reaccion_clasificado']])

# Crea los nombres de las nuevas columnas
encoded_columns = encoder.get_feature_names_out(['Reaccion_clasificado'])

# Crea un DataFrame con los datos codificados
encoded_df = pd.DataFrame(encoded_data, columns=encoded_columns)

# Combina el DataFrame original con el DataFrame codificado
datos_numéricos = pd.concat([datos, encoded_df], axis=1)

# Elimina la columna correspondiente a 'missing' si existe
if 'Reaccion_clasificado_missing' in datos_numéricos.columns:
    datos_numéricos.drop(columns=['Reaccion_clasificado_missing'], inplace=True)

# Elimina la columna original 'Reaccion_clasificado'
datos_numéricos.drop(columns=['Reaccion_clasificado'], inplace=True)

datos_numéricos.head()
```

Género y Cuando

```python
# Map values for 'Género' and 'Cuando' columns
gender_map = {'H': 0, 'M': 1}
when_map = {'A': 1, 'D': 0}

datos_numéricos['Género'] = datos_numéricos['Género'].map(gender_map)
datos_numéricos['Cuando'] = datos_numéricos['Cuando'].map(when_map)



# Display the updated DataFrame to the user
datos_numéricos.head()
```
