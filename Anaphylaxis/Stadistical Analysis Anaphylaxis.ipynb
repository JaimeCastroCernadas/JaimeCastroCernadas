```python
# Crear columnas de ratio para los biomarcadores especificados
data['THBS1_ratio'] = data['THBS1 Agudo'] / data['THBS1 Basal']
data['APOH_ratio'] = data['APOAH Agudo'] / data['APOH Basal']
data['APOA2_ratio'] = data['APOA2 Agudo'] / data['APOA2 Basal']
data['Triptasa_ratio'] = data['Triptasa Aguda'] / data['Triptasa Basal']
data['A2-AP_ratio'] = data['A2-AP Agudo'] / data['A2-AP Basal']

# Mostrar las primeras filas del DataFrame modificado para verificar las nuevas columnas
data[['THBS1_ratio', 'APOH_ratio', 'APOA2_ratio', 'Triptasa_ratio', 'A2-AP_ratio']].head()
```

```python
# Calculando la matriz de correlación para todo el DataFrame modificado
correlation_matrix = data.corr()

# Mostrar la matriz de correlación
correlation_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurando el tamaño del gráfico
plt.figure(figsize=(15, 10))

# Creando el heatmap con anotaciones de los valores de correlación
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)

# Añadiendo títulos y etiquetas
plt.title('Heatmap de Correlaciones entre Variables')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import scipy.stats as stats

shapiro_results = []

for col in data:
    if data[col].dtype in ['int64', 'float64']:  # Asegurarse de que la columna es numérica
        stat, p = stats.shapiro(data[col].dropna())  # La prueba Shapiro-Wilk
        # Determinar si la distribución es normal
        normal = "Sí" if p > 0.05 else "No"
        shapiro_results.append({'Variable': col, 'Estadístico': stat, 'p-value': p, 'Distribución Normal': normal})

# Crear un DataFrame para mostrar los resultados
shapiro_df_data = pd.DataFrame(shapiro_results)
shapiro_df_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Preparando los datos para el boxplot
data_to_plot = [
    data['THBS1_1_ratio'].dropna(),
    data['THBS1_2_ratio'].dropna(),
    data['THBS1_3_ratio'].dropna(),
    data['THBS1_ratio'].dropna()  # Datos totales
]

# Etiquetas para el gráfico
labels = ['Grado 1', 'Grado 2', 'Grado 3', 'Total']

# Creando el boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_to_plot)
plt.xticks(range(4), labels)

# Añadiendo puntos individuales como scatter
for i, d in enumerate(data_to_plot):
    y = d
    x = np.random.normal(i, 0.04, size=len(y))
    plt.plot(x, y, 'r.', alpha=0.4)

# Realizar prueba de ANOVA
anova_result = f_oneway(*data_to_plot)

# Añadiendo el resultado de ANOVA al gráfico
plt.figtext(0.99, 0.01, f'ANOVA p-value: {anova_result.pvalue:.3f}', horizontalalignment='right')

plt.title('Comparación de Ratios de THBS1 por Grado y Total')
plt.ylabel('Ratio Agudo/Basal THBS1')
plt.show()
```

```python
import pandas as pd
import scipy.stats as stats

# Suponemos que 'data' es tu DataFrame y que las columnas numéricas son 'Edad', 'Triptasa Aguda', 'Triptasa Basal', etc.
# Ejecutaremos la prueba de Shapiro-Wilk para cada una de estas columnas
numeric_columns = ['Edad', 'Triptasa_ratio',"Género_codificado","Cuando_codificado","Reaccion_codificada","THBS1_ratio","APOA2_ratio","APOH_ratio","A2-AP_ratio"]  # Añade más columnas según sea necesario
shapiro_results = []

for col in numeric_columns:
    if data[col].dtype in ['int64', 'float64']:  # Asegurarse de que la columna es numérica
        stat, p = stats.shapiro(data[col].dropna())  # La prueba Shapiro-Wilk
        # Determinar si la distribución es normal
        normal = "Sí" if p > 0.05 else "No"
        shapiro_results.append({'Variable': col, 'Estadístico': stat, 'p-value': p, 'Distribución Normal': normal})

# Crear un DataFrame para mostrar los resultados
shapiro_df = pd.DataFrame(shapiro_results)
shapiro_df
```
