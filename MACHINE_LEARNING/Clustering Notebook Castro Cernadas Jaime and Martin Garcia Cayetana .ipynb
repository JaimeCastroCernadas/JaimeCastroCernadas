sin pipeline

Preprocess data with OneHotEncoder and RobustScaler pipeline

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.pipeline import Pipeline
# Define the preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        # Apply OneHotEncoder to the specified categorical columns
        ('onehot', OneHotEncoder(sparse_output=False), ['U95019_s_at', 'HG3729-HT3999_f_at']),
        # Apply RobustScaler to the remaining columns (numerical features)
        ('scaler', RobustScaler(quantile_range=(25.0, 75.0)), list(data.drop(['U95019_s_at', 'HG3729-HT3999_f_at'], axis=1).columns))
    ],
    remainder='passthrough'  # Ensure any columns not specified are kept as-is
)

# Create the complete pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)
])

# Fit and transform the data
processed_data = pipeline.fit_transform(data)
```

Train_test_split

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
train_data, test_data = train_test_split(processed_data, test_size=0.2, random_state=42)

print(train_data.shape)
print(test_data.shape)
```

HIERARCHICAL CLUSTERING

```python
#Define function for the dendogram plotting.
import numpy as np
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram

def plot_dendrogram(model, **kwargs):
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    dendrogram(linkage_matrix, **kwargs)
```

Initial

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

# Fit the data to the model
model = AgglomerativeClustering(n_clusters=3,compute_distances=True)
model = model.fit(train_data)

# Using model labels as the cluster labels
labels = model.labels_
#Calculate the silhouette score.
score = silhouette_score(train_data, labels)

plt.title("Hierarchical Clustering Dendrogram")
#Call the plot_dendrogram function.
plot_dendrogram(model, truncate_mode="level", p=3)
plt.xlabel("Number of points in node (or index of point if no parenthesis)")
plt.show()
print("Silhouette Score:", score)
```

Find the optimal parameters

```python
from sklearn.model_selection import ParameterGrid

# Define the parameter range and linkage types.
param_grid = {
    'n_clusters': [2, 3, 4, 5, 6],
    'linkage': ['ward', 'complete', 'average', 'single'],
}

best_score = -1
best_params = None
best_model = None

# Perform the gridsearch for the parameters.
for params in ParameterGrid(param_grid):
    model = AgglomerativeClustering(**params, compute_distances=True)
    model.fit(train_data)

    # Access the labels from the fitted model
    labels = model.labels_

    # Evaluate the model using silhouette score
    score = silhouette_score(train_data, labels)

    if score > best_score:
        best_score = score
        best_params = params
        best_model = model

# Print the best model and its parameters
print("Mejores parámetros:", best_params)
print("Mejor Silhouette Score:", best_score)

# Visualize the dendrogram of the best model
plt.title("Hierarchical Clustering Dendrogram (Best Model)")
plot_dendrogram(best_model, truncate_mode="level", p=3)
plt.xlabel("Number of points in node (or index of point if no parenthesis)")
plt.show()
```

Validate output for the test data

```python
# Apply the model on the test data
test_labels = best_model.fit_predict(test_data)

# Evaluate the model on the test data using silhouette score
silhouette_test_score = silhouette_score(test_data, test_labels)

# Print the silhouette score for the test data
print(f"Silhouette Score (Test Data): {silhouette_test_score}")

# Plot the dendrogram for the best model on the test data.
plt.title("Hierarchical Clustering Dendrogram on Test data (Best Model)")
plot_dendrogram(best_model, truncate_mode="level", p=3)
plt.xlabel("Number of points in node (or index of point if no parenthesis)")
plt.show()
```

*Partitional Clustering: K-means*

```python
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score


# Fit the train data for  K-means model. K++ is used as it is the default method.
kmeans= KMeans(n_clusters=2, random_state=0)
kmeans.fit(train_data)

# Get the labels and cluster centers
train_data_labels= kmeans.labels_  # Assign cluster labels to the training DataFrame
centroids = kmeans.cluster_centers_ #Get the centroids coordinates.

# Plotting the results for K-means.
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(train_data[:, -1], train_data[:, -2], c=train_data_labels,)
plt.scatter(centroids[:, -1], centroids[:, -2], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-means Clustering with K-means++ Initialization')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

#Get the shilouette score, because with too many features the visualization is impractical.
silhouette_avg = silhouette_score(train_data, kmeans.labels_)
print(f'Silhouette Score: {silhouette_avg}')

#Davies-Bouldin Index
db_index = davies_bouldin_score(train_data, kmeans.labels_)
print(f'Davies-Bouldin Index: {db_index}')

```

```python
from sklearn.metrics import silhouette_samples
from sklearn.metrics.pairwise import pairwise_distances


# Set the range of K values
K_values = range(2, 11)  # Try K from 2 to 10
distortions = []
silhouette_scores = []

for k in K_values:
    # Initialize KMeans with K++
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(train_data)

    # Distortion calculation (sum of squared distances to the nearest cluster center)
    distortion = np.sum(np.min(pairwise_distances(train_data, kmeans.cluster_centers_, metric='euclidean')**2, axis=1))
    distortions.append(distortion)

    # Calculate the silhouette score
    score = silhouette_score(train_data, kmeans.labels_)
    silhouette_scores.append(score)

# Plotting the Distortion and Silhouette Scores
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the distortion
ax1.set_xlabel('Number of Clusters (K)')
ax1.set_ylabel('Distortion', color='tab:blue')
ax1.plot(K_values, distortions, color='tab:blue', marker='o', label='Distortion')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis to plot silhouette scores
ax2 = ax1.twinx()
ax2.set_ylabel('Silhouette Score', color='tab:green')
ax2.plot(K_values, silhouette_scores, color='tab:green', marker='s', label='Silhouette Score')
ax2.tick_params(axis='y', labelcolor='tab:green')

plt.title('Distortion and Silhouette Scores for Different K Values')
plt.grid(True)
plt.show()

# Print the best silhouette score and corresponding distortion
best_k = K_values[np.argmax(silhouette_scores)]  # The K with the highest silhouette score
print(f"Best K based on silhouette score: {best_k}")
print("Mean silhouette score:", silhouette_scores[np.argmax(silhouette_scores)])


```

```python
# Fit the best model using the training data
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(train_data)

# Get the labels and cluster centers
train_data_labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plotting the results for K-means on training data
plt.figure(figsize=(12, 5))

# Training data plot
plt.subplot(1, 2, 1)
plt.scatter(train_data[:, -1], train_data[:, -2], c=train_data_labels)
plt.scatter(centroids[:, -1], centroids[:, -2], c='red', s=200, alpha=0.75, marker='X')
plt.title('Initial K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Get the silhouette score for the training data
silhouette_avg = silhouette_score(train_data, kmeans.labels_)
print(f'Silhouette Score (Training Data): {silhouette_avg}')

# Davies-Bouldin Index for training data
db_index = davies_bouldin_score(train_data, kmeans.labels_)
print(f'Davies-Bouldin Index (Training Data): {db_index}')

# Predict the clusters for the test data
# Use trained KMeans model to predict the labels for the test data
test_data_labels = kmeans.predict(test_data)

# Plotting the test data clusters prediction (same centroids from the train data)
plt.subplot(1, 2, 2)
plt.scatter(test_data[:, -1], test_data[:, -2], c=test_data_labels)
plt.scatter(centroids[:, -1], centroids[:, -2], c='red', s=200, alpha=0.75, marker='X')
plt.title('Test Data Prediction with optimal K ')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.show()

# Print the silhouette score for the test data
silhouette_test = silhouette_score(test_data, test_data_labels)
print(f'Silhouette Score (Test Data): {silhouette_test}')

# Davies-Bouldin Index for test data
db_index_test = davies_bouldin_score(test_data, test_data_labels)
print(f'Davies-Bouldin Index (Test Data): {db_index_test}')
```

**DBSCAN** Initial model

```python
from sklearn.cluster import DBSCAN

# Define DBSCAN with fixed parameters
dbscan = DBSCAN(eps=0.5, min_samples=5)

# Fit DBSCAN to the data
labels = dbscan.fit_predict(train_data)

# Get the number of clusters (excluding noise points labeled as -1)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print(f"Number of clusters: {n_clusters}")

# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(train_data[:, -1], train_data[:, -2], c=labels, cmap='viridis', alpha=0.6)
plt.title(f"DBSCAN Clustering with eps=0.5 and min_samples=5")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Optimal parameters search for DBSCAN

```python
# Grid search parameters for DBSCAN
eps_values = np.linspace(0.1, 2.0, num=10)
min_samples_values = [3, 5, 10, 15, 20]

# Initialize variables to store the best parameters and silhouette score
best_score = -1
best_params = {}

# Perform grid search over the parameter grid
for eps in eps_values:
    for min_samples in min_samples_values:
        # Initialize DBSCAN with current parameters
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)

        # Fit DBSCAN and predict cluster labels
        labels = dbscan.fit_predict(train_data)

        # Check if the clustering resulted in more than 1 cluster (excluding noise labeled as -1)
        unique_labels = set(labels)
        if len(unique_labels) > 1:  # Ensure there is more than one cluster
            score = silhouette_score(train_data, labels)
            print(f"eps: {eps:.2f}, min_samples: {min_samples}, Silhouette Score: {score:.4f}")

            # Track the best parameters based on silhouette score
            if score > best_score:
                best_score = score
                best_params = {'eps': eps, 'min_samples': min_samples}

# Output the best parameters and the associated score
if best_params:
    print("\nBest Parameters:", best_params)
    print("Best Silhouette Score:", best_score)

    # Fit the DBSCAN model with the best parameters
    best_dbscan = DBSCAN(eps=best_params['eps'], min_samples=best_params['min_samples'])
    best_labels = best_dbscan.fit_predict(train_data)

    # Plot the results
    plt.figure(figsize=(8, 6))
    plt.scatter(train_data[:, 0], train_data[:, 1], c=best_labels, cmap='viridis', alpha=0.6)
    plt.title(f"DBSCAN Clustering with eps={best_params['eps']} and min_samples={best_params['min_samples']}")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
else:
    print("No valid clusters found during grid search.")
```

Initial Gaussian Mixture Model

```python
from sklearn.mixture import GaussianMixture

# Fit the initial GMM model with the train data
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(train_data)

# Predict the labels (clusters) for the dataset
labels = gmm.predict(train_data)

# Get the cluster means (centroids) and covariances
means = gmm.means_
covariances = gmm.covariances_

# Calculate BIC score for the model
bic_score = gmm.bic(train_data)
print(f'BIC Score: {bic_score}')

# Plot the data points and the predicted clusters
plt.figure(figsize=(8, 6))
plt.scatter(train_data[:, -1], train_data[:, -2], c=labels, cmap='viridis', alpha=0.6)
plt.scatter(means[:, -1], means[:, -2], c='red', s=200, marker='X', label='Centroids')
plt.title(f'Initial GMM Clustering with {3} Components')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

# Calculate the silhouette score to evaluate clustering performance
silhouette_avg = silhouette_score(train_data, labels)
print(f'Silhouette Score: {silhouette_avg}')
```

Optimized parameters

```python

# Define parameter grid
param_grid = {
    'n_components': [2, 3, 4, 5, 6],
    'covariance_type': ['full', 'tied', 'diag', 'spherical'],  # Covariance types to explore
}

# Initialize variables to track the best model
best_bic = np.inf  # Start with a very high BIC score to ensure the first model is better
best_params = None
best_gmm = None
best_labels = None

# Perform grid search manually
for params in ParameterGrid(param_grid):
    # Create the GMM model
    gmm = GaussianMixture(**params, random_state=42)

    # Fit the model to the data
    gmm.fit(train_data)

    # Calculate BIC for this model
    bic = gmm.bic(train_data)

    # Calculate the number of clusters
    labels = gmm.predict(train_data)

    # If this model has a better BIC , update the best model
    if bic < best_bic:
        best_bic = bic
        best_params = params
        best_gmm = gmm
        best_labels = labels

    # Print the BIC score for this combination of parameters
    print(f"n_components: {params['n_components']}, covariance_type: {params['covariance_type']}, BIC: {bic}")

# Output the best model and its parameters
print(f"\nBest Parameters: {best_params}")
print(f"Best BIC: {best_bic}")

# Predict the clusters using the best GMM model
best_labels = best_gmm.predict(train_data)

# Get the cluster means (centroids) and covariances
means = best_gmm.means_

# Plot the data points and the predicted clusters
plt.figure(figsize=(8, 6))
plt.scatter(train_data[:, -1], train_data[:, -2], c=best_labels, cmap='viridis', alpha=0.6)
plt.scatter(means[:, -1], means[:, -2], c='red', s=200, marker='X', label='Centroids')
plt.title(f'Optimized GMM Clustering with {best_params["n_components"]} Components')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

# Calculate the silhouette score to evaluate clustering performance
silhouette_avg = silhouette_score(train_data, best_labels)
print(f'Silhouette Score: {silhouette_avg}')
```
