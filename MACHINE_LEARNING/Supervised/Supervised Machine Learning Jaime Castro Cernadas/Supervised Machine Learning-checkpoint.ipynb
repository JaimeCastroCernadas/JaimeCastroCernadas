# Data split

First, I will split the data in train and test.

```python
from sklearn.model_selection import train_test_split

#Create the X (features) and Y (target) datasets
X = data_clean.drop(columns=['Target'])
y = data_clean['Target']

#Split into train and test. Why do I use stratify=y? Why test_size=0.3?
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42,stratify=y)

print("Size of training-evaluation data:", X_train.shape)
print("Size of test data:", X_test.shape)
```

# Pipeline

```python
from sklearn.model_selection import GridSearchCV, LeaveOneOut, cross_val_predict
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report, accuracy_score

BEST_MODEL = None

#Define the function for the hyperparameter tuning
def hyperparameter_tuning(X_train, y_train, model, param_grid, cv):
    """
    Perform hyperparameter tuning using GridSearchCV with the provided model and parameter grid.

    Parameters:
    - X_train: Training features.
    - y_train: Training labels.
    - model: A machine learning model.
    - param_grid: Parameter grid for tuning.
    - cv: Cross-validation strategy .

    Returns:
    - best_model: The best model after hyperparameter tuning.
    - best_params: Best hyperparameters.
    - best_score: Best score achieved on validation (accuracy).
    """
    #Create the gridsearch
    grid_search = GridSearchCV(
        estimator=model,
        param_grid=param_grid,
        cv=cv,
        scoring='accuracy',
        verbose=1
    )

    grid_search.fit(X_train, y_train)

    print("Best model is:", grid_search.best_estimator_)
    print("Best parameters are:", grid_search.best_params_)
    print("Best accuracy score in model validation:", grid_search.best_score_)

     # Store the best model in the global variable
    global BEST_MODEL
    BEST_MODEL = grid_search.best_estimator_


    return grid_search.best_estimator_, grid_search.best_params_, grid_search.best_score_

#Define the function for the training data performance report
def training_prediction(X_train, y_train, best_model, cv):
    """
        This function creates the accuracy and classification report 
        for the training and prediction of the whole training data

        Parameters:
        - X_train: Training features.
        - y_train: Training labels.
        - best_model: Tuned machine learning model.
        - cv: Cross-validation strategy.

        Returns:
        -train_classification_report: Classification report from the best model with the training data.

    """
    # Predicting on the training data with cross-validation.
    y_train_pred = cross_val_predict(best_model, X_train, y_train, cv=cv)

    # Generate classification report for the training data.
    train_classification_report = classification_report(y_train, y_train_pred, target_names=["NR", "R"])

    # Output results
    print("Classification Report for Training Data with crossvalidation:\n", train_classification_report)

#Define the function for the test data performance report
def test_prediction_and_matrix(X_test, y_test, best_model, cv):
    """
    This function creates the accuracy and classification report 
    for the test data prediction.

    Parameters:
    - X_test: Test features.
    - y_test: Test labels.
    - best_model: Tuned machine learning model.
    - cv: Number of cross-validation folds or strategy.

    Returns:
    -test_classification_report: Classification report from the best model with the test data.
    -test_accuracy: Accuracy from the best model with the test data.

    """
    
    # Predicting on the test data with cross-validation 
    y_test_pred = best_model.predict(X_test)

    # Generate classification report for the test data.
    test_classification_report = classification_report(y_test, y_test_pred, target_names=["NR", "R"])
    test_accuracy= accuracy_score(y_test,y_test_pred)

    # Output results
    print("Best model accuracy (Test):", test_accuracy )
    print(f"Classification Report for Test Data\n", test_classification_report)

    #Plot the confussion matrix
    cm = confusion_matrix(y_test, y_test_pred, labels=["NR", "R"])

    # Plot the confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["NR", "R"])
    disp.plot(cmap='Blues')

#Pipeline that combines the 3 previous functions.
def model_pipeline(X_train, y_train, X_test, y_test, model, param_grid, cv):
    """
    Pipeline to perform hyperparameter tuning, training prediction, and test prediction.

    Parameters:
    - X_train: Training features.
    - y_train: Training labels.
    - X_test: Test features.
    - y_test: Test labels.
    - model: A machine learning model.
    - param_grid: Parameter grid for tuning.
    - cv: Cross-validation strategy.

    Returns:
    - best_model: The best model after hyperparameter tuning.
    - best_params: Best hyperparameters.
    - best_score: Best score achieved on validation (accuracy).
    """
    #  Hyperparameter tuning
    best_model, best_params, best_score = hyperparameter_tuning(X_train, y_train, model, param_grid, cv)
    
    
    # Training prediction and performance report
    training_prediction(X_train, y_train, best_model, cv)
    
    #  Test prediction and performance report
    test_prediction_and_matrix(X_test, y_test, best_model, cv)
    
```
